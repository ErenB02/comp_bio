### Split the dataset in 80% training and 20% test.

```python
df.sample(len(df))
```

```python
train_size = round(len(df)*0.8)

# Shuffle the dataset
df = df.sample(len(df))

df_train = df.iloc[:train_size]
df_test = df.iloc[train_size:]
```
