```python
#Task 1

sugar_wght = float(input("What is the weight of the sugar in grams ? "))
flour_wght = float(input("What is the weight of the flour in grams ? "))
fruit_wght = float(input("What is the weight of the fruit in grams ? "))
total_wght = sugar_wght + flour_wght + fruit_wght
print(f"{total_wght:.2f}")
```

```python
#Task 2
sugar_cal = sugar_wght * 3.87
flour_cal = flour_wght * 3.64
fruit_cal = fruit_wght * 0.52
calories = sugar_cal + flour_cal + fruit_cal
print(f"{calories:.1f}")
```

```python
#Task 3
if fruit_wght >= 40.0:
  print("1 portion of your 5 A Day")
else:
    print("Not 1 portion of your 5 A Day")
```

```python
#Task 4

if 0 >= calories <= 200:
  print("The snack is Dietetic")
elif 201 >= calories <= 400:
  print("The snack is Normal")
else:
  print("The snack is Highly caloric")
```
